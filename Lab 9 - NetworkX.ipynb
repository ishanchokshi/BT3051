```python
DG = nx.DiGraph()
print(nx.info(DG))
nx.draw(DG,with_labels=True)
plt.show()
```

```python
DG = nx.DiGraph()
DG.add_edge(2,1) #Adds edge in the direction from 2 to 1
print(nx.info(DG))
nx.draw(DG,with_labels=True)
plt.show()
```

```python
DG = nx.DiGraph()
DG.add_edge(2,1) #Adds edge in the direction from 2 to 1
DG.add_edge(1,2)
DG.add_edge(1,4)
DG.add_edge(1,3)
DG.add_edge(2,4)
print(nx.info(DG))
nx.draw(DG,with_labels=True)
plt.show()
```
