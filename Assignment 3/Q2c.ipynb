```python
# no libraries are allowed

""" Add your functions here """

def optimal_investment_stocks_selection(price_2011, price_2021, investment_amount, n):
	"""
	(list of int), (list of int), (int), (int) -> (list of int)

	>>> optimal_investment_stocks_selection([1000,1000,1800,100,1200], [3900, 1300, 4700, 120, 4500], 3000, 5)
	    [3, 0, 0, 0, 0]
	"""
	
	if len(price_2011) != n or len(price_2021) != n: # comparison should be done with length
		raise("Incorrect Input!")
	
	"""
	Let w[i] be the number of stocks bought of the ith company
	Constraint: (sum over i, w[i] * price_2011[i]) = investment_amount (As we have to invest all our money)
	Objective: maximise: sum over i, w[i] * (price_2021[i] - price_2011[i]) 
	"""

	profit = [price_2021[i] - price_2011[i] for i in range(n)]
	dp = [0] * (investment_amount + 1) # dp[x] denotes the maximal profit given total initial amount is 'x'
	bought = [-1] * (investment_amount + 1) # bought[x] denotes which stock was bought given amount 'x'
	for amount in range(investment_amount + 1):
		for stock in range(n):
			if (price_2011[stock] > amount):
				continue
			bought_stock_profit = profit[stock] + dp[amount - price_2011[stock]]
			if bought_stock_profit > dp[amount]:
				dp[amount] = bought_stock_profit
				bought[amount] = stock

	w = [0] * n
	current_amount = investment_amount
	while current_amount != 0 and bought[current_amount] != -1:
		w[bought[current_amount]] += 1
		current_amount -= price_2011[bought[current_amount]]

	return w

print(optimal_investment_stocks_selection([1000,1000,1800,100], [3900, 1300, 4700, 120, 4500], 3000, 5))
```
