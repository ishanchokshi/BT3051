### HAMM

```python
def hamming_distance (s,t):
    distance = 0;
    for i in range(len(s)):
        if(s[i]!=t[i]):
            distance = distance+1
        
    return distance
```

```python
length = int(input("Enter length: "))
s = []
t = []
for i in range(length):
    ele1 = input("Enter nucleotides for s: ")
    ele2 = input("Enter nucleotides for t: ")
    s.append(ele1)
    t.append(ele2)

Hamming_distance = hamming_distance(s,t)
print("Hamming distance is: ",Hamming_distance)
```
